# Data Visualization

```python
tw['Tweet Location'].value_counts()
```

```python
#Getting top_10 Tweet Location
tw['Tweet Location'].value_counts()[0:10]
```

```python
#plot bar graph
plt.xlabel("Tweet Location")
plt.ylabel("count")
plt.title('Distribution of Tweet Location')
plt.bar(tw['Tweet Location'].value_counts().keys()[0:10],tw['Tweet Location'].value_counts()[0:10])
plt.show()
```

```python
tw['Tweet Language'].value_counts()
```

```python
#plot bar graph
plt.xlabel("Tweet Language")
plt.ylabel("count")
plt.title('Distribution of Tweet Language')
c=['r','m','y']
plt.bar(list(tw['Tweet Language'].value_counts().keys()),list(tw['Tweet Language'].value_counts()),color=c,label=['English','German','Polish'])
plt.legend()
plt.show()
```

```python
#Plot pie chart
plt.title("Pie Chart")
plt.pie(list(tw['Tweet Language'].value_counts()),labels=list(tw['Tweet Language'].value_counts().keys()),autopct="%0.2f%%")
plt.show()
```

```python
tw['Tweet Type'].value_counts()
```

```python
#plot bar graph
plt.xlabel("Tweet Type")
plt.ylabel("count")
plt.title('Distribution of Tweet Type')
c=['r','m','y']
plt.bar(list(tw['Tweet Type'].value_counts().keys()),list(tw['Tweet Type'].value_counts()),color=c,label=['ReTweet','Tweet','Reply'])
plt.legend()
plt.show()
```
